```python
#find the list of count percentage for filling the null values into the ratio wise
lst=[]
for i in vaccine['REGION'].value_counts().values:
    lst.append(round(i/(len(vaccine)-vaccine['REGION'].isnull().sum()),2))
```

```python
lst#The values are divided into equal portion into percentage
```

```python
#Filling the null values of region column
vaccine['REGION'] = vaccine['REGION'].fillna(pd.Series(np.random.choice(['Flanders', 'Brussels', 'Wallonia','Ostbelgien'], 
                                                      p=lst, size=len(vaccine))))
```

```python
vaccine.isnull().sum()#after the filling the null values there are no null values in the region column
```

```python
vaccine['REGION'].value_counts(normalize=True)#It shows the count percentage of each category after filling null values
```

```python
vaccine_sex_values = vaccine['SEX'].value_counts().values#It shows the category wise count of sex column
```

```python
null_values_of_sex = vaccine['SEX'].isnull().sum()
null_values_of_sex#null values count of sex column
```

```python
#find the list of count percentage for filling the null values into the ratio wise
lst_of_sex = []
for i in vaccine_sex_values:
    lst_of_sex.append(round(i/(len(vaccine)-null_values_of_sex),5))
```

```python
lst_of_sex#The values are divided into equal portion into percentage
```

```python
#Filling the null values of sex column
vaccine['SEX'] = vaccine['SEX'].fillna(pd.Series(np.random.choice(['F','M'], 
                                                      p=lst_of_sex, size=len(vaccine))))
```

```python
vaccine['SEX'].value_counts(normalize=True)#It shows the count percentage of each category after filling null values
```
